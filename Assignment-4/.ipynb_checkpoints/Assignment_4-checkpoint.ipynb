```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")


# 1. Read the provided CSV file
df = pd.read_csv("data.csv")

# 2. Show the basic statistical description about the data
print(df.describe())

# 3. Check if the data has null values and replace them with the mean
if df.isnull().values.any():
    df = df.fillna(df.mean())

# 4. Select at least two columns and aggregate the data using: min, max, count, mean
agg_df = df[['Duration', 'Calories']].agg(['min', 'max', 'count', 'mean'])
print(agg_df)

# 5. Filter the dataframe to select the rows with calories values between 500 and 1000
filter_df1 = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print(filter_df1)

# 6. Filter the dataframe to select the rows with calories values > 500 and pulse < 100
filter_df2 = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print(filter_df2)

# 7. Create a new “df_modified” dataframe that contains all the columns from df except for “Maxpulse”
df_modified = df.drop('Maxpulse', axis=1)
print("df_modified:\n", df_modified)

# 8. Delete the “Maxpulse” column from the main df dataframe
df.drop('Maxpulse', axis=1, inplace=True)

# 9. Convert the datatype of Calories column to int datatype
#df['Calories'] = df['Calories'].astype(int)
print(df.dtypes)
df=df.astype({'Calories':'int'})
print(df.dtypes)

# 10. Using pandas create a scatter plot for the two columns (Duration and Calories)
df.plot.scatter(x='Duration', y='Calories')

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Load the Glass dataset
glass_data = pd.read_csv('glass.csv')

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(glass_data.iloc[:, :-1], glass_data['Type'], test_size=0.25, random_state=0)

# Initialize the Gaussian Naive Bayes model and fit the data
nb = GaussianNB()
nb.fit(X_train, y_train)

# Predict on the test data and calculate accuracy
y_pred = nb.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Naïve bayes Accuracy: ",accuracy)

# Print classification report
print(classification_report(y_test, y_pred))




#SVM Algorithm
from sklearn.svm import SVC

# Initialize the SVM model and fit the data
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)

# Predict on the test data and calculate accuracy
y_pred = svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Linear SVM Accuracy: ",accuracy)

# Print classification report
print(classification_report(y_test, y_pred, zero_division=1))

#Visualisation
import matplotlib.pyplot as plt

# Create a scatter plot with refractive index on the x axis and the glass type on the y axis
plt.scatter(glass_data['RI'], glass_data['Type'], label='Refractive Index')
plt.scatter(glass_data['Mg'], glass_data['Type'], label='Mg')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Glass Type')
plt.title('Correlation Between Features and Glass Type')
# Display legend
plt.legend()
plt.show()
```

```python
#Linear SVM algorithm gives better accuracy when compared with Naive bayes algorithm:
SVMs are different from other Naive Bayes algorithm because of the way they choose the decision boundary that maximizes 
the distance from the nearest data points of all the classes. The decision boundary created by
SVMs is called the maximum margin classifier or the maximum margin hyper plane.
```
